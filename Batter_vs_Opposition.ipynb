```python
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import numpy as np 
import os
import pandas as pd
```

```python
df1 = pd.read_csv('D:/CRICKET_IPL/fantasy_cricket/archive/deliveries.csv')
df1.dataframeName = 'deliveries.csv'
```

```python
df1.head(5)
```

```python
final_df = df1[(df1["batter"] == "S Dhawan") & (df1["bowling_team"] == "Royal Challengers Bangalore")]
final_df
```

```python
#total_stats
total_runs = final_df["batsman_runs"].sum()
print(total_runs)

total_innings = len(pd.unique(final_df['match_id']))
print(total_innings)

total_outs = final_df[final_df["player_dismissed"] == 'S Dhawan']
print(len(total_outs))

average = total_runs / len(total_outs)
print(average)

strike_rate = total_runs / len(final_df) * 100
print(strike_rate)

dots = final_df[final_df["batsman_runs"] == 0]
dot_percentage = len(dots)/len(final_df) * 100

print(dot_percentage)

boundaries = final_df[final_df["batsman_runs"] == 4]
sixes =  final_df[final_df["batsman_runs"] == 6]
boundary_percentage = len(boundaries) + len(sixes) / len(final_df) * 100
print(boundary_percentage)
```
