```python
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import numpy as np 
import os
import pandas as pd
```

```python
df = pd.read_csv('archive/IPL_Matches_2008_2022.csv')
df.dataframeName = 'matches.csv'
```

```python
season_subset = df[(df["Season"] == 2022 )]
season_subset
```

```python
team_subset = season_subset[season_subset["Team2"] == 'Rajasthan Royals'] 

match_id_list = team_subset["ID"].to_numpy()

match_id_list.tolist()
print(match_id_list)
```

```python
df1 = pd.read_csv('archive/IPL_Ball_by_Ball_2008_2022.csv')
df1.dataframeName = 'deliveries.csv'
df1
```

```python
final_df = []
for i in range(len(df1)):
    if df1.iat[i,0] in match_id_list:
        final_df.append(df1.iloc[i])
```

```python
final_df
```

```python
final_df = pd.DataFrame(final_df)
final_df = final_df[final_df["innings"]==1]
```
